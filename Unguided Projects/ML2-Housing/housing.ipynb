# Normalizing some columns(even though we dont really need it here for tree based xgboost or random forest)

```python
from sklearn.preprocessing import MinMaxScaler
df5 = df4
nor_cln = ['LotFrontage','FullBath','GarageCars','TotRmsAbvGrd','OverallQual','LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF','GrLivArea', 'GarageYrBlt', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF']
scaler = MinMaxScaler()
df5[nor_cln] = scaler.fit_transform(df5[nor_cln])

```

```python
df5.head(5)
```

```python
#xgboost(use lasso regression),random forest,ensembling,
```

# MODEL CREATION

```python
x  = df5.drop(columns='SalePrice',axis=1)
y = df5['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_squared_error
import xgboost as xgb 
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
from math import sqrt

```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
#xgbooster with lasso
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', alpha=1, random_state=42)
xgb_model.fit(x_train, y_train)
y_pred_xgb = xgb_model.predict(x_test)
```

```python
y_pred_xgb
```

```python
#prediction with Random forest regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
y_pred_rf = rf_model.predict(x_test)
```

```python
y_pred_rf
```

```python
#Ensembling RF and XGB(may use stacking or blending)
y_pred_ensemble = (y_pred_xgb + y_pred_rf) / 2
```

```python
y_pred_ensemble
```

```python
#using squareroot of mse as its been told root mean-squared error has been removed fro scikit-learn
rmse_xgb = sqrt(mean_squared_error(y_test, y_pred_xgb))
rmse_rf = sqrt(mean_squared_error(y_test, y_pred_rf))
rmse_ensemble = sqrt(mean_squared_error(y_test, y_pred_ensemble))
print('xgb result=',rmse_xgb,',','rf_result=',',',rmse_rf,',','ensemble_result=',rmse_ensemble)
```

# FEATURE IMPORTANCE

```python
# Plot feature importances from inbuild xbg feature importance
xgb.plot_importance(xgb_model, importance_type='weight', title='Feature Importance by Weight')
plt.show()

xgb.plot_importance(xgb_model, importance_type='gain', title='Feature Importance by Gain')
plt.show()

xgb.plot_importance(xgb_model, importance_type='cover', title='Feature Importance by Cover')
plt.show()
```

```python
#importing feature importance
importances = xgb_model.feature_importances_

# Sort features by importance
sorted_idx = importances.argsort()
plt.figure(figsize=(10, 8))
plt.barh(x.columns[sorted_idx], importances[sorted_idx])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance from XGBoost Model')
plt.show()
```
